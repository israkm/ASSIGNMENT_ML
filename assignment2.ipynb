```python
# Calculate the most common winning age range
age_bins = [10, 15, 20, 25, 30, 35, 40, 45, 50, 55, 60, 65, 70]
age_labels = ['10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40-44', '45-49', '50-54', '55-59', '60-64', '65-69']

# Use .loc to avoid SettingWithCopyWarning
medal_winners.loc[:, 'Age_Range'] = pd.cut(medal_winners['Age'], bins=age_bins, labels=age_labels, right=False)
most_common_age_range = medal_winners['Age_Range'].value_counts().idxmax()

print(f"The most common winning age range for medal winners is: {most_common_age_range} years.")
```
